```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x_in = np.linspace(-5.0, 5.0, num=50)
y_in = np.linspace(-5.0, 5.0, num=50)
x, y = np.meshgrid(x_in, y_in)
z = x ** 2 + y ** 2

fig, ax = plt.subplots(1, 1)
cp = ax.contour(x, y, z, colors="blue", linestyles="dashed")
plt.show()
```
